```python
import pandas as pd 
import numpy as np
```

```python
data1 = pd.read_csv('Experiment-Data/data1.csv')
data1.head()
```

```python
extracted_data = data1[['stim_duration', 'stim', 'coded_response', 'exp_stage','trial_num']]
extracted_data = extracted_data.dropna(subset=['trial_num'])
```

```python
def tastiest(data):
    filtered_data = data[(data['exp_stage'] == 'taste_rating') & (data['coded_response'] == 2)]
    if not filtered_data.empty:
        return filtered_data['stim'].values
    else:
        return None

def untastiest(data):
    filtered_data = data[(data['exp_stage'] == 'taste_rating') & (data['coded_response'] == -2)]
    if not filtered_data.empty:
        return filtered_data['stim'].values
    else:
        return None

def healthiest(data):
    filtered_data = data[(data['exp_stage'] == 'health_rating') & (data['coded_response'] == 2)]
    if not filtered_data.empty:
        return filtered_data['stim'].values
    else:
        return None

def unhealthiest(data):
    filtered_data = data[(data['exp_stage'] == 'health_rating') & (data['coded_response'] == -2)]
    if not filtered_data.empty:
        return filtered_data['stim'].values
    else:
        return None

def confusing_items(data):
    filtered_data = data[(data['exp_stage'] == 'decision') & (data['stim_duration'] < 4000)]
    max_duration = filtered_data['stim_duration'].max()
    result = filtered_data[filtered_data['stim_duration'] == max_duration]
    if not result.empty:
        return result['stim'].values
    else:
        return None

def unavoidable_items(data):
    filtered_data = data[data['exp_stage'] == 'decision']
    min_duration = filtered_data['stim_duration'].min()
    result = filtered_data[filtered_data['stim_duration'] == min_duration]
    if not result.empty:
        return result['stim'].values
    else:
        return None
```

```python
print(tastiest(extracted_data))
print(untastiest(extracted_data))
print(healthiest(extracted_data))
print(unhealthiest(extracted_data))
print(confusing_items(extracted_data))
print(unavoidable_items(extracted_data))
```

```python
def healthyDF(data):
    filtered_data = data[(data['exp_stage'] == 'health_rating') & (data['coded_response'] > 0)]
    stims = filtered_data['stim'].values
    decision_data = data[(data['stim'].isin(stims)) & (data['exp_stage'] == 'decision')]
    return decision_data

def unhealthyDF(data):
    filtered_data = data[(data['exp_stage'] == 'health_rating') & (data['coded_response'] < 0)]
    stims = filtered_data['stim'].values
    decision_data = data[(data['stim'].isin(stims)) & (data['exp_stage'] == 'decision')]
    return decision_data

def tastyDF(data):
    filtered_data = data[(data['exp_stage'] == 'taste_rating') & (data['coded_response'] > 0)]
    stims = filtered_data['stim'].values
    decision_data = data[(data['stim'].isin(stims)) & (data['exp_stage'] == 'decision')]
    return decision_data

def untastyDF(data):
    filtered_data = data[(data['exp_stage'] == 'taste_rating') & (data['coded_response'] < 0)]
    stims = filtered_data['stim'].values
    decision_data = data[(data['stim'].isin(stims)) & (data['exp_stage'] == 'decision')]
    return decision_data
```
